```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

model_id = "microsoft/Phi-3-mini-4k-instruct"

print("正在使用原生模式加载模型（跳过远程脚本）...")

# 1. 加载模型
# 核心改变：trust_remote_code=False
# 这会强制使用 transformers 库自带的、经过测试的稳定代码
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=False  # ✨ 核心修复：不再下载并运行远程的 modeling_phi3.py
)

# 2. 加载分词器
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 3. 创建推理流水线
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=True,
    temperature=0.7
)

# 4. 测试生成
messages = [{"role": "user", "content": "Create a funny joke about chickens."}]
print("\n" + "="*30 + "\n✨ AI 生成结果：\n")
output = generator(messages)
print(output[0]["generated_text"])
```
