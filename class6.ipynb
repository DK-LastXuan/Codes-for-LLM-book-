```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_id = "microsoft/Phi-3-mini-4k-instruct"

# 加载模型
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=False,  # 更新库后，建议设为 False 或直接删掉这行
    # attn_implementation="eager" # 如果你没有安装 flash_attn，可能会有警告，加上这个可以消除警告
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

print("模型加载成功！")
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False,
)
```

```python
messages = [
    {"role": "user", "content": "Tell me what is T1 in LOL?"}
]
```

```python
output = pipe(messages, do_sample=True, temperature=1)
print(output[0]["generated_text"])
```

```python
# 使用较高的top_p
output = pipe(messages, do_sample=True, top_p=1)
print(output[0]["generated_text"])
```

```python
one_shot_prompt = [
    {
        "role": "user",
        "content": "A 'Gigamuru' is a type of Japanese musical instrument. An example of a sentence that uses the word Gigamuru is:"
    },
    {
        "role": "assistant",
        "content": "I have a Gigamuru that my uncle gave me as a gift. I love to play it at home."
    },
    {
        "role": "user",
        "content": "To 'screeg' something is to swing a sword at it. An example of a sentence that uses the word screeg is:"
    }
]
print(tokenizer.apply_chat_template(one_shot_prompt, tokenize=False))

# 生成输出
outputs = pipe(one_shot_prompt)
print(outputs[0]["generated_text"])
```

```python
!pip install llama-cpp-python --extra-index-url https://abetlen.github.io/llama-cpp-python/whl/cu121
!pip install huggingface-hub
from llama_cpp import Llama

# 加载模型，将所有层放到GPU上运行（n_gpu_layers=-1）
llm = Llama.from_pretrained(
    repo_id="microsoft/Phi-3-mini-4k-instruct-gguf",  # 替换为实际的Hugging Face仓库ID
    filename="*fp16.gguf",  # 替换为模型文件名
    n_gpu_layers=-1,  # 所有层都在GPU上运行
    n_ctx=4096,  # 上下文长度，根据模型和需求调整
    verbose=False
)
```

```python
# 生成输出
output = llm.create_chat_completion(
    messages=[
        {"role": "user", "content": "Create a warrior for an RPG in JSON format."},
    ],
    response_format={"type": "json_object"},
    temperature=0,
)['choices'][0]['message']["content"]
```

```python
import json

# 格式化为JSON
json_output = json.dumps(json.loads(output), indent=4)
print(json_output)
```
